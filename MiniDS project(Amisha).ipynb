```python
import pandas as pd
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sInputFileName='C:/olympic/dataset_olympics.csv'
InputData=pd.read_csv(sInputFileName,encoding="latin-1")
```

```python
print(InputData)
```

```python
InputData['Sex'].value_counts()
```

```python
plt.pie(InputData['Sex'].value_counts(), labels=['Male','Female'], autopct=('%1.0f%%'), pctdistance=0.85, explode=(0.02,0.02))
centre_circle = plt.Circle((0, 0), 0.55, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
```

```python
print('Number of participants per year')
YearData=InputData
InputData['Year'].value_counts()
```

```python
sns.countplot(data=InputData, x="Year")
plt.title("Number of participants per year")
plt.xticks(rotation=92)
plt.show()
```

```python
InputData['Medal'].value_counts()
```

```python
sns.countplot(data=InputData, x="Year", hue="Medal")
plt.title("Medals distribution per years")
plt.xticks(rotation=92)
plt.show()
```
